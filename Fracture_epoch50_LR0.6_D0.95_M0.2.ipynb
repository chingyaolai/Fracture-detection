```python
from __future__ import division, print_function
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
plt.rcParams['image.cmap'] = 'gray'
np.random.seed(98765)
```

```python
from tf_unet import image_util
from tf_unet import unet
from tf_unet import util
```

```python
prediction_path = "./prediction_3layers_feature32_epoch50_LR0.6_D0.95_M0.2"
output_path = "./trained_3layers_feature32_epoch50_LR0.6_D0.95_M0.2"
training_data = image_util.ImageDataProvider("./data_train/*.tif")
```

```python
i=11 #number of images
x_train, y_train = training_data(i)
train_y = y_train[...,1].flatten()
print(train_y.max())
train_x = x_train[...,0].flatten()
print(train_x.max())
```

```python
net = unet.Unet(layers=3, features_root=32, channels=1, n_class=2)
trainer = unet.Trainer(net, optimizer="momentum", opt_kwargs=dict(momentum=0.2,learning_rate=0.6,decay_rate=0.95))
#trainer = unet.Trainer(net)
path = trainer.train(training_data, output_path, training_iters=32, epochs=50, 
                     display_step=2, prediction_path=prediction_path)
```

```python
import tensorflow as tf
epochs = 50
acc=[]
acc_=[]
loss=[]
loss_=[]
lr=[]
lr_=[]

import glob
eventfile = glob.glob(output_path+'/*.yao')[0]

for event in tf.train.summary_iterator(eventfile):
#    for value in event.summary.value:
#        print(value.tag)
#        if value.HasField('simple_value'):
#            print(value.simple_value)
    for v in event.summary.value:
        if v.tag == 'accuracy':
            acc.append(v.simple_value)
        if v.tag == 'loss':
            loss.append(v.simple_value)
        if v.tag == 'learning_rate':
            lr.append(v.simple_value)

iter = np.arange(0, len(acc))
iter_=[]
epoch = np.arange(0, epochs)

for i in np.arange(0,len(acc),int(len(acc)/epochs)):
    acc_.append(np.mean(acc[i:i+int(len(acc)/epochs)]))
    loss_.append(np.mean(loss[i:i+int(len(loss)/epochs)]))
    lr_.append(np.mean(lr[i:i+int(len(lr)/epochs)]))
    iter_.append(np.mean(iter[i:i+int(len(iter)/epochs)]))

plt.figure()
plt.plot(iter, acc, lw=1, linestyle = '--')
plt.plot(iter_, acc_)
plt.xlabel('iterations')
plt.ylabel('accuracy')
plt.ylim([0.9,1])
plt.figure()
plt.plot(iter, loss, lw=1, linestyle = '--')
plt.plot(iter_, loss_)
plt.xlabel('iterations')
plt.ylabel('loss')
plt.ylim([0,0.2])
print('loss in the final epoch is '+ str(loss_[len(epoch)-1]))
plt.figure()
plt.plot(iter, lr, lw=1, linestyle = '--')
plt.plot(iter_, lr_)
plt.xlabel('iterations')
plt.ylabel('learning rate')
plt.ylim([0,1])
```

### Test performance

```python
#prepare test data
test_data = image_util.ImageDataProvider("./data_train/*.tif")
i=11 #number of images
x_test, y_test = test_data(i)

net = unet.Unet(layers=3, features_root=32, channels=1, n_class=2)
prediction = net.predict(output_path+"/model.ckpt", x_test)

print(x_test.shape)
print(y_test.shape)
print(prediction.shape)
```

```python
from sklearn.utils import class_weight
from sklearn.metrics import confusion_matrix, precision_score, recall_score, log_loss, f1_score, classification_report
from sklearn.metrics import roc_curve, auc, roc_auc_score, precision_recall_curve, average_precision_score, accuracy_score

#test_y = y_test[...,20:1020,20:1020,1].flatten()
test_y = y_test[...,20:980,20:980,1].flatten()
test_y_predict = prediction[...,1].flatten()
predict_max = max(test_y_predict)
test_y_predict = test_y_predict/predict_max

print(max(test_y))
print(max(test_y_predict))
print(test_y.shape)
print(test_y_predict.shape)

print(classification_report(test_y, (test_y_predict > 0.4)))
```

```python
print(classification_report(test_y, (test_y_predict > 0.82)))
```

```python
fpr, tpr, thresholds = roc_curve(test_y, test_y_predict)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw=lw, label = 'ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color = 'navy', lw=lw, linestyle = '--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc = 'lower right')
plt.show()
print('AUC = ',roc_auc)

precision, recall, thresholds = precision_recall_curve(test_y, test_y_predict)
plt.figure()
plt.plot(recall, precision)
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
